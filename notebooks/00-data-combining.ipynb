```python


# Step 1: Import Libraries

import pandas as pd


# Step 2: Load Both Datasets

df_us = pd.read_csv("/kaggle/input/data-combine/USvideos.csv", encoding='utf-8')
df_in = pd.read_csv("/kaggle/input/data-combine/INvideos.csv", encoding='utf-8')

print("âœ… USA dataset shape:", df_us.shape)
print("âœ… India dataset shape:", df_in.shape)


# Step 3: Add Country Labels

df_us['country'] = 'USA'
df_in['country'] = 'India'


#Step 4: Combine Both Files

df_combined = pd.concat([df_us, df_in], ignore_index=True)
print("âœ… Combined dataset shape:", df_combined.shape)


# Step 5: Remove Duplicates

df_combined.drop_duplicates(subset=['video_id', 'country'], inplace=True)
print("ðŸ§¹ After removing duplicates:", df_combined.shape)


# Step 6: Save Combined Dataset

output_path = "/kaggle/working/US_IN_videos_combined.csv"
df_combined.to_csv(output_path, index=False, encoding='utf-8')

print(f"âœ… Combined dataset saved successfully at: {output_path}")


#Step 7: Quick Preview

df_combined.head()
```
